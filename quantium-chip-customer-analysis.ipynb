```python
data = pd.merge(transactions, customers, on='LYLTY_CARD_NBR', how='left')
```

```python
data.head()
```

```python
data['TOT_SALES'].sum()
```

```python
sales_by_segment = data.groupby(['LIFESTAGE','PREMIUM_CUSTOMER'])['TOT_SALES'].sum()

```

```python
sales_by_segment
```

```python
qty_by_segment = data.groupby(['LIFESTAGE','PREMIUM_CUSTOMER'])['PROD_QTY'].sum()

qty_by_segment
```

```python
avg_qty_per_customer = data.groupby(['LIFESTAGE','PREMIUM_CUSTOMER'])['PROD_QTY'].mean()

avg_qty_per_customer
```

```python
data['PRICE_PER_PACKET'] = data['TOT_SALES'] / data['PROD_QTY']
```

```python
price_by_segment = data.groupby(['LIFESTAGE','PREMIUM_CUSTOMER'])['PRICE_PER_PACKET'].mean()

price_by_segment
```

```python
brand_sales = data.groupby('BRAND')['TOT_SALES'].sum().sort_values(ascending=False)

brand_sales
```

```python
brand_sales.head(10).plot(kind='bar', figsize=(10,5))

plt.title("Top Chip Brands by Sales")
plt.ylabel("Total Sales")
plt.show()
```

```python
pack_sales = data.groupby('PACK_SIZE')['TOT_SALES'].sum().sort_values(ascending=False)

pack_sales
```

```python
pack_sales.head(10).plot(kind='bar', figsize=(10,5))
plt.title("Chips sales by pack size")
plt.ylabel("Total Sales")
plt.xlabel("Pack size(g)")
plt.show()
```

```python
segment_sales = data.groupby(['LIFESTAGE','PREMIUM_CUSTOMER'])['TOT_SALES'].sum().sort_values(ascending=False)

segment_sales

segment_sales.head(10).plot(kind='bar', figsize=(12,6))
plt.title("Top Customer Segments Driving Chip Sales")
plt.ylabel("Total Sales")
plt.show()
```
